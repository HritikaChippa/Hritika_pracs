```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/averaging_filter.bmp",0)
padded_img=np.pad(img,((1,1),(1,1)),mode='constant')
filter=np.zeros_like(img)
kernel=np.ones((3,3),dtype=np.float32)/9.0
```

```python
m,n=img.shape
for i in range (m):
  for j in range (n):
    filter[i,j]=np.sum(kernel*padded_img[i:i+3,j:j+3])
output=np.uint8(filter)
cv2_imshow(img)
cv2_imshow(output)
```

```python
img=cv2.imread("/content/averaging_filter.bmp",0)
padded_img=np.pad(img,((1,1),(1,1)),mode='constant')
filter=np.zeros_like(img)
kernel=np.array([[1,2,1],[2,4,2],[1,2,1]])/16.0
```

```python
m,n=img.shape
for i  in range (m):
  for j in range (n):
    weighted_sum=0.0;
    kernel_sum=0.0;
    for k in range (-1,2):
      for l in range (-1,2):
        pixel_val=padded_img[i+k+1,j+l+1]
        weight=kernel[k+1,l+1]
        weighted_sum+=pixel_val*weight
        kernel_sum+=weight
        filter[i,j]=weighted_sum/kernel_sum

output=np.uint8(filter)
cv2_imshow(img)
cv2_imshow(output)
```

```python
img=cv2.imread("/content/median.bmp",0)
filter=np.zeros_like(img)
k_size=3
p_size=k_size
m,n=img.shape
for i in range(p_size,m-p_size):
  for j in range(p_size,n-p_size):
    window=img[i-p_size:i+p_size+1,j-p_size:j+p_size+1]
    med=np.median(window)
    filter[i,j]=med

output=np.uint8(filter)
cv2_imshow(img)
cv2_imshow(output)
```
