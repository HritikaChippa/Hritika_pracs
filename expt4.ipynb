```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/hist1.bmp",0)
cv2_imshow(img)
hist=cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
img=cv2.imread("/content/hist4.bmp",0)
cv2_imshow(img)
hist=cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
img=cv2.imread("/content/hist4.bmp",0)
cv2_imshow(img)
hist,bins=np.histogram(img.ravel(),256,[0,256])
plt.plot(hist)
plt.show()
```

```python
img=cv2.imread("/content/hist1.bmp",0)
cv2_imshow(img)
hist=cv2.calcHist([img],[0],None,[256],[0,256])


plt.plot(hist)
plt.title("Hist 1")
plt.axis('off')
plt.show()

img2=cv2.equalizeHist(img)
hist2=cv2.calcHist([img2],[0],None,[256],[0,256])

plt.plot(hist2)
plt.title("Hist 1")
plt.axis('off')

plt.show()
```
