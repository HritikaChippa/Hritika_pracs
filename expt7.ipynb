```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/Edge detectn.bmp",0)
img.shape
```

```python
img=cv2.resize(img,(500,500),fx=0.7,fy=0.7)
```

```python
cv2_imshow(img)
```

```python
def roberts(img):
  k1=np.array([[1,0],[0,-1]],dtype=np.float32)
  k2=np.array([[0,1],[-1,0]],dtype=np.float32)
  gx=cv2.filter2D(img,-1,k1)
  gy=cv2.filter2D(img,-1,k2)
  mag=np.abs(gx)+np.abs(gy)
  mag=np.uint8(mag)
  return gx,gy,mag

gx,gy,mag=roberts(img)
plt.figure(figsize=(15,8))
plt.subplot(1,3,1)
plt.imshow(gx,cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(gy,cmap='gray')
plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(mag,cmap='gray')
plt.axis('off')
```
