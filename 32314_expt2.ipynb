```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/flower2.JPG")
cv2_imshow(img)
```

```python
color=('b','g','r')
for i,col in enumerate(color):
  hist=cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(hist,color=col)
plt.show()
```

```python
img_neg=255-img
cv2_imshow(img_neg)
```

```python
color=('b','g','r')
for i,col in enumerate(color):
  hist=cv2.calcHist([img_neg],[i],None,[256],[0,256])
  plt.plot(hist,color=col)
plt.show()
```

```python
c=255/np.log(1+np.max(img))
log_img=c*(np.log(1+img))
s=np.array(log_img,dtype=np.uint8)
cv2_imshow(s)
```

```python
gamma1=0.2
img1=255*(img/255)**gamma1
cv2_imshow(img1)
```

```python
img=cv2.imread("/content/fl1.bmp",0)
w,h =img.shape
z=img.copy()
cv2_imshow(img)
for x in range (w-1):
  for y in range (h-1):
    z[x,y]=img[x,h-y-1]
cv2_imshow(z)
```
